## Train model

````
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
````

````
learn.fit_one_cycle(4)
````

````
learn.save('stage-1')
````

````
learn.unfreeze()
````

````
learn.lr_find()
````

Now we have to use the learning rate finder plot to identify the strongest downward slope that's sticking arround for quite a wild. 

````
learn.recorder.plot()
````

````
# If the plot is not showing try to give a start and end learning rate
# learn.lr_find(start_lr=1e-5, end_lr=1e-1)
# As we can se in the upper plot, if I don't pass start and end parameters to lr_find, the plot increasees from (approx.) 1e-03 to 1e-02
# That's why we pass the following parameters
learn.lr_find(start_lr=1e-06, end_lr=1e-03)

````

````
learn.recorder.plot()
````

Now we are able to view the behavior with more detail. 
We have to identify the period in which the learning rate deacreases the fastest for a quite long period of time, and that will be our start_lr parameter. For end_lr we use 1e-4 or 3e-4, this is a rule of thumb.


````
learn.fit_one_cycle(2, max_lr=slice(6e-05,3e-4))
````

````
learn.save('stage-2')
````

## Cleaning Up
Some of our top losses aren't due to bad performance by our model. There are images in our data set that shouldn't be.
This models are pretty good at dealing with a moderate ammount of noisy data. The problem would occur if the data isn't randomly noisy, but biased noisy. If ather cleaning the data it is %0.01 better, it's fine, but still it's a good idea just to make sure that there isn't too much noise in my data in case it is biased.

Using the ImageCleaner widget from fastai.widgets we can prune our top losses, removing photos that don't belong.

````
from fastai.widgets import *
````

First we need to get the file paths from our top_losses. We can do this with `.from_toplosses`. We then feed the top losses indexes and corresponding dataset to `ImageCleaner`.

Notice that the widget will not delete images directly from disk but it will create a new csv file `cleaned.csv` from where you can create a new ImageDataBunch with the corrected labels to continue training your model.

````
app = Starlette()
````

````
@app.route("/classify-url", methods=["GET"])
async def classify_url(request):
    bytes = await get_bytes(request.query_params["url"])
    img = open_image(BytesIO(bytes))
    _,_,losses = learner.predict(img)
    return JSONResponse({
        "predictions": sorted(
            zip(cat_learner.data.classes, map(float, losses)),
            key=lambda p: p[1],
            reverse=True
        )
    })
````
