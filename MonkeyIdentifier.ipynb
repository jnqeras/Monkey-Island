###Let's create a validation set!

Whenever you create a data bunch, if you don't have a separate training and validation set, then you can just say the training set is in the current folder (i.e. . because by default, it looks in a folder called train) and I want you to set aside 20% of the data. So this is going to create a validation set for you automatically and randomly. You'll see that whenever I create a validation set randomly, I always set my random seed to something fixed beforehand. This means that every time I run this code, I'll get the same validation set. In general, I'm not a fan of making my machine learning experiments reproducible (i.e. ensuring I get exactly the same results every time). The randomness is to me a really important part of finding out your is solution stable and it is going to work each time you run it. But what is important is that you always have the same validation set. Otherwise when you are trying to decide has this hyper parameter change improved my model but you've got a different set of data you are testing it on, then you don't know maybe that set of data just happens to be a bit easier. So that's why I always set the random seed here.

##Trying to inprove.

###Things that can go wrong:

Most of the time things will train fine with the defaults values.


*   But sometimes that won't work, in that case, there's not much you really need to tune 

*  Most likely are
** High learnign rate:
> If the learning rate is too high, the **validation loss will be very high (usually is under 1)**. Also if the valid_loss is too high, then the learning rate (max_lr) will high to (regardless of how many epochs it does). If this happens, there is no way to undo it, you have to go back and create your neural net again and fit from scratch with a lower learning rate.
** Low learning rate:
>If the learning rate it too low, the error_rate gets better, but very slowly **(you can plot the error_rate with learn.recorder.plot_losses())**. The other thing that will happen if the learning rate is too low is that **the training loss will be higher than the validation loss (and you never will want this in your model)**.
** High number of epochs:
>If the number of epochs is too high: it will produce overfitting (despite what you've heard, it's hard to overfit with deep leaning). The only thing that tells you **if you are overfitting** is that **the error rate improves for a while and then starts getting worse again** (get's smaller for a while and then starts to increase again).
** Low number of epochs:
>If the number of epochs is too low, the trainging loss is much higher than the validation loss (similarly to what happens when the learning rate is too low).
